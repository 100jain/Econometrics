**1.   BERNOULLI TRIAL**

```
import matplotlib.pyplot as plt 
import seaborn as sns
from statsmodels.distributions.empirical_distribution import ECDF

# Import the bernoulli object from scipy.stats
from scipy.stats import bernoulli

# Set the random seed to reproduce the results
import numpy as np
np.random.seed(10)

# Simulate thousand coin flip with 30% chance of getting heads
sample=bernoulli.rvs(p=0.30, size=1000)
print(sample)

# Plot the sample
sns.distplot(sample)
plt.show()

#Plotting CDF
ecdf= ECDF(sample)

# plot the cdf
plt.plot(ecdf.x, ecdf.y)
plt.show()
```


**2.   BINOMIAL DISTRIBUTION**

*Bernaulli trail - flipping of one coin*

*Binomial distrbution - repeated bernaulli trails*

```
# Import the bernoulli object from scipy.stats
from scipy.stats import binom

# Simulate 20 trials of 10 coin flips 
sample= binom.rvs(n=10, p=0.30, size=20)
print(sample)

# Plot the sample
sns.distplot(sample)
plt.show()

#Plotting CDF
ecdf= ECDF(sample)

# plot the cdf
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

**3.   NORMAL DISTRIBUTION**

```
# Import the normal object from scipy.stats
from scipy.stats import norm

# Create the sample using norm.rvs(), loc = mean and scale = std. deviation
sample = norm.rvs(loc=3.15, scale=1.5, size=10000, random_state=13)

print(sample)

# Plot the sample
sns.distplot(sample)
plt.show()

#Plotting CDF
ecdf= ECDF(sample)

# plot the cdf
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

```
#Probability density function
pdf=norm.pdf(0,loc=0,scale=1)
print(pdf)

#Cummulation distribution function
cdf=norm.cdf(0,loc=0,scale=1)
print(cdf)
```

**4.   POISSON DISTRIBUTION**

```
# Import the object from scipy.stats
from scipy.stats import poisson

# Create the sample using poisson.rvs(), mu=lamda
sample = poisson.rvs(mu=7, size=10000, random_state=13)

print(sample)

# Plot the sample
sns.distplot(sample)
plt.show()

#Plotting CDF
ecdf= ECDF(sample)

# plot the cdf
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

**5.   GEOMETRIC DISTRIBUTION**

```
# Import the object from scipy.stats
from scipy.stats import geom

# Create the sample p = probability of success
sample = geom.rvs(p=0.3, size=10000, random_state=13)

print(sample)

# Plot the sample
sns.distplot(sample)
plt.show()

#Plotting CDF
ecdf= ECDF(sample)

# plot the cdf
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

**6.   P Values from cumulative distribution functions**

```
import numpy as np
np.set_printoptions(precision=4)  # print arrays to 4 decimal places
import matplotlib.pyplot as plt
import scipy.stats
t_dist = scipy.stats.t(20)
t_values = np.linspace(-4, 4, 1000)
# Show relationship of PDF and CDF for three example t values.
example_values = (-1.5, 0, 1.5)
pdf_values = t_dist.pdf(t_values)
cdf_values = t_dist.cdf(t_values)
fill_color = (0, 0, 0, 0.1)  # Light gray in RGBA format.
line_color = (0, 0, 0, 0.5)  # Medium gray in RGBA format.
fig, axes = plt.subplots(2, len(example_values), figsize=(10, 6))
for i, x in enumerate(example_values):
    cdf_ax, pdf_ax = axes[:, i]
    cdf_ax.plot(t_values, cdf_values)
    pdf_ax.plot(t_values, pdf_values)
    # Fill area at and to the left of x.
    pdf_ax.fill_between(t_values, pdf_values,
                        where=t_values <= x,
                        color=fill_color)
    pd = t_dist.pdf(x)  # Probability density at this value.
    # Line showing position of x on x-axis of PDF plot.
    pdf_ax.plot([x, x],
                [0, pd], color=line_color)
    cd = t_dist.cdf(x)  # Cumulative distribution value for this x.
    # Lines showing x and CDF value on CDF plot.
    x_ax_min = cdf_ax.axis()[0]  # x position of y axis on plot.
    cdf_ax.plot([x, x, x_ax_min],
                [0, cd, cd], color=line_color)
    cdf_ax.set_title('x = {:.1f}, area = {:.2f}'.format(x, cd))
    # Hide top and right axis lines and ticks to reduce clutter.
    for ax in (cdf_ax, pdf_ax):
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.yaxis.set_ticks_position('left')
        ax.xaxis.set_ticks_position('bottom')
```
